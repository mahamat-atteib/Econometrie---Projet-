**Question 1. Énoncer l'hypothèse fondamentale pour que les estimateurs OLS soient non biaisés**

Le modèle linéaire s'écrit : $y = X\beta + \varepsilon$

L'hypothèse fondamentale qui permet aux estimateurs des moindres carrés ordinaires (OLS) d'être non biaisés est  <span style="font-size:15px; color:blue; font-weight:bold"> l'exogénéité </span> **des variables explicatives**.   
Cette hypothèse est exprimée mathématiquement comme suit:
   
   $$E(\epsilon | X) = 0$$
   
   Cela signifie que l'espérance conditionnelle des termes d'erreur, étant donné les variables explicatives, est égale à zéro. En d'autres termes, il ne doit pas y avoir de corrélation entre les variables indépendantes et les erreurs du modèle. Cela garantit que les variations dans les variables indépendantes ne sont pas systématiquement liées aux variations non expliquées dans la variable dépendante.
   
La moyenne du bruit est aussi nulle : $E(\varepsilon) = 0$



**Question 2. Montrez que sous cette hypothèse, les estimateurs OLS sont effectivement non biaisés**

Pour montrer que les estimateurs OLS sont non biaisés sous l'hypothèse d'exogénéité, considérons le modèle de régression linéaire simple:
$$Y = \beta_0 + \beta_1X + \epsilon$$

### Estimations OLS
Les estimateurs OLS pour $\beta_0$ et $\beta_1\$ sont donnés par les formules :

$$\hat{\beta}_1 = \frac{\sum_{i=1}^n (X_i - \overline{X})(Y_i - \overline{Y})}{\sum_{i=1}^n (X_i - \overline{X})^2}$$
$$\hat{\beta}_0 = \overline{Y} - \hat{\beta}_1 \overline{X}$$

Nous pouvons réexaminer l'expression de $\hat{\beta}_1$ et montrer comment les erreurs individuelles $\epsilon_i$ contribuent à cet estimateur, ce qui peut aider à comprendre le rôle de l'exogénéité de manière plus explicite.

L'estimateur OLS pour $\beta_1$ est donné par :
$$\hat{\beta}_1 = \frac{\sum_{i=1}^n (X_i - \overline{X})(Y_i - \overline{Y})}{\sum_{i=1}^n (X_i - \overline{X})^2}$$

En substituant $Y_i = \beta_0 + \beta_1 X_i + \epsilon_i$ dans cette expression, nous avons :
$$\hat{\beta}_1 = \frac{\sum_{i=1}^n (X_i - \overline{X})(\beta_0 + \beta_1 X_i + \epsilon_i - \overline{Y})}{\sum_{i=1}^n (X_i - \overline{X})^2}$$

Simplifiant, où $\overline{Y} = \beta_0 + \beta_1 \overline{X} + \overline{\epsilon}$, et les termes constants $\beta_0$ et $\overline{Y}$ s'annulent :
$$\hat{\beta}_1 = \frac{\sum_{i=1}^n (X_i - \overline{X})(\beta_1 (X_i - \overline{X}) + \epsilon_i)}{\sum_{i=1}^n (X_i - \overline{X})^2}$$

D'ici, nous pouvons réarranger pour illustrer le terme correctif mentionné :
$$\hat{\beta}_1 = \beta_1 + \frac{\sum_{i=1}^n (X_i - \overline{X}) \epsilon_i}{\sum_{i=1}^n (X_i - \overline{X})^2}$$

Ce terme additionnel montre comment les erreurs individuelles $\epsilon_i$ pondérées par leur distance à la moyenne de $X$ affectent l'estimation de $\beta_1$. L'exogénéité, $E(\epsilon_i \mid X_i) = 0$, garantit que l'espérance de ce terme additionnel est zéro, ce qui montre que $\hat{\beta}_1\$ est non biaisé :

$$E\left(\frac{\sum_{i=1}^n (X_i - \overline{X}) \epsilon_i}{\sum_{i=1}^n (X_i - \overline{X})^2}\right) = 0$$

Cela confirme que, sous l'hypothèse d'exogénéité, les variations dues aux erreurs dans la relation entre $X$ et $Y$ ne biaisent pas l'estimation de $\beta_1$ en moyenne.


### Preuve de non-biais

Pour prouver que ces estimateurs sont non biaisés, nous devons montrer que l'espérance des estimateurs est égale aux vrais paramètres.

- **Espérance de $\hat{\beta}_1$**:
  
  $$
E(\hat{\beta}_1) = E\left(\frac{\sum_{i=1}^n (X_i - \overline{X})(Y_i - \overline{Y})}{\sum_{i=1}^n (X_i - \overline{X})^2}\right) = E\left(\beta_1 + \frac{\sum_{i=1}^n (X_i - \overline{X}) \epsilon_i}{\sum_{i=1}^n (X_i - \overline{X})^2}\right)
$$


  on en déduit bien que
  Substituant $Y_i$ par $\beta_0 + \beta_1X_i + \epsilon_i$, et en utilisant l'hypothèse $E(\epsilon | X) = 0$, cela simplifie à :
  $$E(\hat{\beta}_1) = \beta_1$$
  $$E(\hat{\beta}_1) - \beta_1 = 0 $$
  car les termes d'erreur $\epsilon_i$ s'annulent en espérance.

- **Espérance de $\hat{\beta}_0$**:
  $$E(\hat{\beta}_0) = E(\overline{Y} - \hat{\beta}_1\overline{X})$$
  $$E(\hat{\beta}_0) = E(\overline{Y}) - E(\hat{\beta}_1)\overline{X}$$
  $$E(\hat{\beta}_0) = \beta_0 + \beta_1\overline{X} - \beta_1\overline{X} = \beta_0$$
  $$E(\hat{\beta}_0) - \beta_0 = 0 $$
  où $\overline{Y} = \beta_0 + \beta_1\overline{X} + \overline{\epsilon}$ et $\overline{\epsilon}$ s'annule en espérance sous notre hypothèse.

Ces démonstrations montrent que, sous l'hypothèse d'exogénéité, les estimateurs OLS $\hat{\beta}_0$ et $\hat{\beta}_1$ sont effectivement **non biaisés**. Cela signifie que, en moyenne, ils atteindront les vraies valeurs des paramètres du modèle.

**Question 4. Expliquez le biais de variable omise avec un exemple du cours**

**Le biais de variable omise** survient lorsqu'une ou plusieurs variables pertinentes ne sont pas incluses dans un modèle statistique, ce qui peut entraîner des estimations erronées de l'effet des variables incluses. Ce biais peut fausser les conclusions d'une étude en attribuant à tort l'effet des variables omises à celles qui sont prises en compte.  
**Use case**: le cours a discuté ce biais à travers l'exemple de l'effet des engrais sur le rendement agricole du soja.  
**Contexte**: dans le cadre de l'agriculture, il est crucial de comprendre les facteurs qui influencent le rendement des cultures pour optimiser l'utilisation des ressources comme les engrais. Cependant, le rendement du soja dépend non seulement de la quantité d'engrais appliquée mais aussi d'autres facteurs tels que la qualité de la terre, le climat, et la présence de parasites.  
**problème de biais de variable omise**: dans l'exemple étudié, l'analyse statistique initiale cherchait à établir une relation entre la quantité d'engrais et le rendement sans prendre en compte la qualité de la terre. Si dans la pratique, les parcelles de meilleure qualité reçoivent plus d'angrais, les resultats de l'étude pourraient indiquer à tort que l'engrais a un effet plus significatif sur le rendement qu'il n'en a réellement. Ce biais de variable omise peut conduire à une surestimation de l'effet de l'engrais et à des décisions suboptimales concernant son utilisation.ation.

Il existe différentes mesures de la multicolinéarité. Dans le cadre ce projet, nous avons opter pour l’approche la plus classique qui consiste à examiner **les facteurs d’inflation de la variance (FIV) ou variance inflation factor (VIF)** en anglais. Les VIF estiment de combien la variance d’un coefficient est augmentée en raison d’une relation linéaire avec d’autres prédicteurs.

```python
# Copie du dataFrame et Ajout d'une constante pour l'intercept
df_copy = df.copy()
df_copy = add_constant(df_copy)

# Exclusion des variables dépendantes pour le calcul du VIF
df_copy = df_copy.drop(['price', 'lprice'], axis=1)

# Calcul du VIF pour chaque variable
vif_data = pd.DataFrame()
vif_data["Variable"] = df_copy.columns
vif_data["VIF"] = [variance_inflation_factor(df_copy.values, i) for i in range(df_copy.shape[1])]

print(vif_data)
```

```python
# Liste des varaibles dépendantes 'price' et 'lprice' qui n'entrent pas dans le calcul du VIF
variables_to_drop = ['price', 'lprice']

# Dropout (exclusion) des variables non nécessaires
df_reduced = df.drop(variables_to_drop, axis=1)

# Ajout d'une constante pour l'intercept
df_reduced = add_constant(df_reduced)

# Calcul du Facteur d'Inflation de la Variance (VIF) pour chaque variable restante
vif_data = pd.DataFrame()
vif_data["Variable"] = df_reduced.columns
vif_data["VIF"] = [variance_inflation_factor(df_reduced.values, i) for i in range(df_reduced.shape[1])]

print(vif_data)
```

**Le problème de multicolinéarité** dans les modèles de régression se produit lorsque deux ou plusieurs prédicteurs (variables indépendantes) dans un modèle de régression sont hautement corrélés entre eux. Cela peut conduire à des coefficients de régression instables, des intervalles de confiance larges, et des difficultés à distinguer l'effet individuel de chaque variable explicative.

**Question 9. Exécuter une régression du prix sur une constante, crime, NOX, chambres, taxes foncières. Commenter l'histogramme des résidus. Interpréter tous les coefficients.**

```python
# Sélection des variables indépendantes et de la variable dépendante 'price'
X = df[['crime', 'nox', 'rooms', 'proptax']]  
y = df['price'] 

# Ajout d'une constante (colonne pour l'intercept)
X = sm.add_constant(X)

# Création du modèle de régression linéaire
model = sm.OLS(y, X).fit()

# Affichage du résumé du modèle
print(model.summary())

# Extraction des résidus
residus = model.resid

# Création de l'histogramme des résidus
plt.hist(residus, bins=30, edgecolor='k', alpha=0.5)
plt.title("Histogramme des résidus de la régression")
plt.xlabel("Résidus")
plt.ylabel("Fréquence")
plt.show()
```

**Question 14. Dans la spécification de la question 9, tester l'hypothèse   $   H_0: \beta_{crime} = \beta_{proptax} $  au niveau de signification de 10% en utilisant la p-valeur du test.**

Pour tester l'hypothèse $H_0: \beta_{\text{crime}} = \beta_{\text{proptax}} $, il est souhaitable de procèder à la recontruction des varaibles du test d'hypothèse en introduisant un nouveau paramètre $\theta_1 = \beta_{\text{crime}} - \beta_{\text{proptax}} $. Cela nous donne une nouvelle variable qui capte la différence entre l'effet de crime et l'effet de proptax. Le test initial est donc équivalent à $ H_0: \theta_1 \!=\! 0 $ &nbsp; <span style="font-size:14px; color:red; font-weight:bold"> vs </span> &nbsp; $ H_1: \theta_1 \!<\! 0$  
Dans le modèle reformulé, nous avons alors :

$\text{price} = \beta_0 + (\theta_1 + \beta_{\text{proptax}}) \times \text{crime} + \beta_{\text{proptax}} \times \text{proptax} + \beta_{\text{nox}} \times \text{nox} + \beta_{\text{rooms}} \times \text{rooms} + u$

Ensuite, nous simplifions cela pour l'ajuster à une régression linéaire :

$\text{price} = \beta_0 + \theta_1 \times \text{crime} + \beta_{\text{proptax}} \times (\text{crime} + \text{proptax}) + \beta_{\text{nox}} \times \text{nox} + \beta_{\text{rooms}} \times \text{rooms} + u$


```python
# Créotion de la nouvelle variable pour la somme de 'crime' et 'proptax'.
df['crime_proptax_sum'] = df['crime'] + df['proptax']

# Sélection des variables pour la régression.
X_1 = df[['crime', 'crime_proptax_sum', 'nox', 'rooms']]
y = df['price']

# Ajout de la constante pour le terme d'intercept.
X_1 = sm.add_constant(X_1)

# Création et entrainement du  modèle de régression linéaire.
model = sm.OLS(y, X_1).fit()

# Affichage des resultats statistiques
print(model.summary())
```

Dans votre modèle, la p-valeur associée à $ \theta_1 $ est de 0.379, ce qui est supérieur que 0.10. Par conséquent, on ne peut pas rejeter l'hypothèse nulle $ H_0$, ce qui signifie que le modèle ne fournit pas suffisamment de preuves pour rejeter l'hypothèse nulle selon laquelle ces deux effets sont équivalents sur le prix des propriétés.


**Question 15. Dans la spécification de la question 9, tester l'hypothèse   &nbsp;  $   H_0: \beta_{NOX} = 0 $  &nbsp; , &nbsp;  $ H_1: \beta_{NOX} = 0 $   &nbsp; au niveau de signification de 10%**

```python
# Modèle non contraint
X_non_contraint = df[['crime', 'nox', 'rooms', 'proptax']]
X_non_contraint = sm.add_constant(X_non_contraint)
model_non_contraint = sm.OLS(y, X_non_contraint)
results_non_contraint = model_non_contraint.fit()
print("Résumé du modèle non contraint:")
print(results_non_contraint.summary())  

# Modèle contraint
X_contraint = df[['crime', 'rooms']]
X_contraint = sm.add_constant(X_contraint)
model_contraint = sm.OLS(y, X_contraint)
results_contraint = model_contraint.fit()
print("Résumé du modèle contraint:")
print(results_contraint.summary())  # Afficher le résumé du modèle contraint

# Calcul des résidus et SSR pour les tests F
residus_non_contraint = results_non_contraint.resid
SSR_non_contraint = residus_non_contraint.T @ residus_non_contraint
residus_contraint = results_contraint.resid
SSR_contraint = residus_contraint.T @ residus_contraint

# Calcul de la statistique F
n = len(y)
k_non_contraint = X_non_contraint.shape[1] - 1
k_contraint = X_contraint.shape[1] - 1
p = k_non_contraint - k_contraint  # Nombre de paramètres contraints
F = ((SSR_contraint - SSR_non_contraint) / p) / (SSR_non_contraint / (n - k_non_contraint - 1))

# Afficher la valeur F et la p-valeur associée
print("Valeur de la statistique F:", F)
from scipy.stats import f
p_value = 1 - f.cdf(F, p, n - k_non_contraint - 1)
print("P-valeur du test F:", p_value)
```

**Conclusion** : Avec une valeur de la statistique de **Fisher F= 19.3113** et une p-valeur extrêmement faible de $8.33 × 10 ^ −9$,  le rejet de l'hypothèse nulle est statistiquement justifié. &nbsp; On rejete l'hypothèse nulle  &nbsp; $ H_0 : \beta_{\text{nox}} = 0$  &nbsp;,  &nbsp;  $\beta_{\text{proptax}} = 0 $  &nbsp; au niveau de signification de 10%.
```python
# PSéparation des données en 2 groupes ( faible niveau de nox et moyen + haut niveau de nox)
low_nox_data = df[df['nox_category'] == 'Low']
medium_high_nox_data = df[df['nox_category'].isin(['Medium', 'High'])]

# Modèle pour le groupe à faible NOx
y_low = low_nox_data['price']
X_low = sm.add_constant(low_nox_data[['crime', 'rooms', 'proptax']])
model_low = sm.OLS(y_low, X_low).fit()
SSR_low = model_low.ssr

# Modèle pour les groupes à NOx moyen et élevé
y_mh = medium_high_nox_data['price']
X_mh = sm.add_constant(medium_high_nox_data[['crime', 'rooms', 'proptax']])
model_mh = sm.OLS(y_mh, X_mh).fit()
SSR_mh = model_mh.ssr

# Modèle global sans séparation
X_global = sm.add_constant(df[['crime', 'rooms', 'proptax']])
y_global = df['price']
model_global = sm.OLS(y_global, X_global).fit()
SSR_global = model_global.ssr

# Nombre total d'observations et de paramètres dans chaque modèle
n = len(df)
k = X_global.shape[1]  # Nombre de prédicteurs incluant la constante

# Test de Chow
F = ((SSR_global - (SSR_low + SSR_mh)) / k) / ((SSR_low + SSR_mh) / (n - 2 * k))
p_value = 1 - f.cdf(F, k, n - 2 * k)

print("Valeur F:", F, "\nP-valeur:", p_value)
```

### Question 25 : Define strict and weak stationarity.

**Stationnarité stricte :** Une série temporelle est considérée comme strictement stationnaire si ses propriétés statistiques telles que la moyenne, la variance, l'autocorrélation, etc. sont invariantes par rapport aux décalages temporels. Cela signifie que la distribution conjointe des observations $(X_t, X_{t+k})$ reste la même, quelque soit la valeur de $t$. La stationnarité stricte exige que tous les moments (moyenne, variance, asymétrie, kurtosis, etc.) et même les propriétés d'ordre supérieur soient identiques à travers toute la série temporelle. C'est une hypothèse forte et souvent pas nécessaire pour les applications pratiques.

**Stationnarité faible :** Une série temporelle est considérée comme faiblement stationnaire (ou stationnaire au sens de la covariance) si elle satisfait les trois conditions suivantes :\
    - La moyenne de la série doit être constante dans le temps.\
    - La variance de la série doit être la même pour les temps $t$.\
    - La covariance entre deux points temporels doit dépendre uniquement de la distance ou du décalage entre les points temporels, et non du moment exact auquel la covariance est calculée.

### Question 26 : Explain ergodicity and state the ergodic theorem. Illustrate with an example.

Une série temporelle ou un processus stochastique est dit **ergodique** si, en prenant un seul échantillon de ce processus sur une longue période, on peut obtenir des informations statistiques (comme la moyenne et la variance) qui sont représentatives de l'ensemble de l'espace des états du processus.

**Théorème ergodique :** D'une manière simple, **le théorème ergodique affirme que pour un processus stochastique stationnaire et ergodique, la moyenne temporelle converge vers la moyenne d'ensemble quand le nombre d'observations tend vers l'infini**. Autrement dit, la moyenne calculée sur une longue série de données d'un processus ergodique est équivalente à la moyenne que l'on obtiendrait en observant simultanément un grand nombre de réalisations de ce processus.

**Exemple illustratif**

Imaginons un jeu de pile ou face avec une pièce équilibrée. Chaque lancer de pièce est un processus stochastique avec une probabilité de 0,5 pour pile et 0,5 pour face. Si nous lançons la pièce un très grand nombre de fois, la fréquence relative de l'apparition de pile devrait converger vers 0,5, qui est la probabilité d'obtenir pile pour un seul lancer.

Soyons précis, supposons que nous lançons la pièce 1000 fois et que nous calculons la fréquence de pile : si nous obtenons 498 piles, la fréquence relative de pile est $\frac{498}{1000} = 0,498$.

**Selon le théorème ergodique, si nous continuons de lancer la pièce un nombre de fois encore plus grand (10 000, 100 000 fois, etc.), la fréquence relative de pile se rapprochera de plus en plus de 0,5**.

### Question 27 : Why do we need both stationarity and ergodicity?

La stationnarité et l'ergodicité sont deux propriétés importantes en analyse des séries temporelles qui jouent des rôles complémentaires dans l'étude et la modélisation des processus stochastiques. Nous avons besoin de ces deux propriétés :

**Stationnarité :**
Consistance temporelle : **La stationnarité assure que les propriétés statistiques du processus (telles que la moyenne, la variance et l'autocorrélation) ne changent pas avec le temps.** Cela simplifie l'analyse et la modélisation car on peut supposer que les lois qui régissent le processus sont constantes dans le temps.

**Prévision et modélisation :** Pour les modèles prédictifs comme les ARMA (AutoRegressive Moving Average), la stationnarité est une condition préalable nécessaire. Sans stationnarité, les prédictions basées sur des données historiques seraient peu fiables car les propriétés sous-jacentes du processus pourraient avoir changé.

**Ergodicité :**
Estimation statistique : **L'ergodicité permet d'utiliser des moyennes temporelles (c'est-à-dire des moyennes calculées sur un seul processus observé sur une longue période) comme estimateurs fiables des moyennes d'ensemble (moyennes calculées sur de nombreuses réalisations du processus).** Cette propriété est essentielle pour assurer que les estimations basées sur un échantillon unique sont représentatives de l'ensemble de la distribution.

**Inférence statistique :** En pratique, nous n'avons souvent accès qu'à un seul échantillon d'un processus (une seule série temporelle). L'ergodicité justifie l'utilisation de cet échantillon unique pour faire des inférences sur le processus global, permettant de tirer des conclusions significatives sur la base de cet échantillon.\
\
**Complémentarité de ces deux propriétés (stationnarité et ergodicité) :**
Efficacité des méthodes statistiques : La stationnarité assure que les paramètres du modèle ne dépendent pas du temps, tandis que l'ergodicité garantit que les estimations faites à partir d'un échantillon unique sont généralisables. Ensemble, ces propriétés permettent une modélisation et une analyse efficaces et robustes des séries temporelles.

### Question 28 : Explain “spurious regression”.

La régression fallacieuse appelée aussi "spurious regression", est un concept important en économétrie, surtout lorsqu'il s'agit de travailler avec des séries temporelles. Ce phénomène se produit lorsque deux séries temporelles non stationnaires sont utilisées dans une régression linéaire, donnant lieu à **des résultats qui peuvent sembler statistiquement valides, mais qui sont en fait trompeurs ou incorrects**.\
\
**Causes de la régression fallacieuse**
**Tendance commune :** Deux variables non stationnaires, souvent caractérisées par une tendance au fil du temps (comme une croissance économique ou une inflation croissante), peuvent sembler corrélées simplement parce qu'elles évoluent toutes deux dans le temps de manière similaire. Cette corrélation est souvent due à la présence de racines unitaires dans les séries, et non à une relation causale entre les variables.\
\
**Corrélation induite :** La corrélation entre les erreurs résiduelles ou entre les variables elles-mêmes peut mener à des statistiques de test trompeuses, faisant paraître les relations entre les variables plus fortes qu'elles ne le sont en réalité.

**Conséquences de la régression fallacieuse**\
**Coefficients trompeurs :** Les estimations des coefficients dans une régression fallacieuse peuvent être biaisées ou inconsistantes, menant à des conclusions erronées sur la relation entre les variables.

**Statistiques de test invalides :**
Les tests statistiques standard (comme le test t de Student pour la significativité des coefficients) peuvent indiquer à tort que les variables sont significativement liées, alors que la "relation" observée peut simplement être due au hasard ou à la présence de tendances dans les données.

**Comment détecter et éviter la régression fallacieuse**\
**Tests de stationnarité :** Avant de réaliser une régression avec des séries temporelles, il est crucial de tester chaque variable pour la stationnarité. Des tests comme le test de Dickey-Fuller augmenté (ADF) peuvent aider à déterminer si une série présente une racine unitaire.

**Cointégration :** Si les variables sont non stationnaires, une approche consiste à tester la cointégration, qui examine si une combinaison linéaire des variables est stationnaire. Si les variables sont cointégrées, cela indique une relation à long terme stable entre elles, malgré leur non stationnarité individuelle.\
**Différenciation :** Une autre méthode consiste à utiliser les premières différences des séries pour rendre les données stationnaires avant la régression.

### Question 30 : Using the original dataset, test the unit root hypothesis for all variables.

À l’aide de l’ensemble de données d’origine, testez l’hypothèse de racine unitaire pour toutes les variables.

- Hypothèse nulle $H_0$       : présence d'une racine unitaire (c'est-à-dire que la série n'est pas stationnaire)
- Hypothèse alternative $H_1$ : absence d'une racine unitaire (la série est stationnaire)

```python
from statsmodels.tsa.stattools import adfuller

# Fonction pour effectuer le test ADF et imprimer les résultats
def adf_test(series, name):
    result = adfuller(series.dropna())  # Suppression des NaN pour permettre le test
    print(f'ADF Statistic for {name}: {result[0]}')
    print(f'p-value for {name}: {result[1]}')
    for key, value in result[4].items():
        print(f'Critical Values for {name}: {key}, {value}')
    print('')

df_filtre = df_filtre.copy()

# Conversion des colonnes en numérique en utilisant .loc pour éviter les avertissements
df_filtre.loc[:, 'real_gdp'] = pd.to_numeric(df_filtre['real_gdp'], errors='coerce')
df_filtre.loc[:, 'unemployment_rate'] = pd.to_numeric(df_filtre['unemployment_rate'], errors='coerce')
df_filtre.loc[:, 'consumer_price_inflation'] = pd.to_numeric(df_filtre['consumer_price_inflation'], errors='coerce')

# Appliquer le test ADF aux colonnes converties
variables_to_test = ['real_gdp', 'unemployment_rate', 'consumer_price_inflation']

for column in variables_to_test:
    print(f"Test ADF pour la colonne : {column}")
    # Maintenant, nous pouvons être sûrs que les colonnes sont de type numérique
    adf_test(df_filtre[column], column)
```

**Seule, la série temporelle "consumer_price_inflation" est stationnaire car sa p-value (égale à 0.001168599408868895) est inférieure à 0.05.**

### Question 32 : Explain the difference between ACF and PACF

La Fonction d'Autocorrélation (ACF) et la Fonction d'Autocorrélation Partielle (PACF) sont toutes deux des outils essentiels utilisés dans l'analyse des séries chronologiques pour décrire la corrélation entre les observations à différents retards temporels. Comprendre leurs différences est essentiel pour modéliser et prévoir efficacement les données de séries chronologiques.

### Fonction d'Autocorrélation (ACF)
- **Ce qu'elle mesure :** La ACF mesure la corrélation entre une série temporelle et ses versions retardées. Elle fournit la corrélation de la série avec elle-même à chaque retard.
- **But :** La ACF est utilisée pour identifier la corrélation totale entre différents points dans une série temporelle, en tenant compte des effets directs et indirects. Cela inclut l'effet combiné des corrélations à tous les retards plus courts qui pourraient intervenir entre deux points.
- **Utilisation :** La ACF est particulièrement utile pour identifier l'ordre des processus de moyenne mobile (MA) dans l'analyse des séries temporelles. Elle aide à comprendre dans quelle mesure les valeurs passées influencent les valeurs futures.

### Fonction d'Autocorrélation Partielle (PACF)
- **Ce qu'elle mesure :** La PACF mesure la corrélation entre une série temporelle et ses versions retardées, mais après avoir éliminé les variations expliquées par les comparaisons intermédiaires. En d'autres termes, elle reflète l'effet direct des observations passées sur la valeur future de la série.
- **But :** La PACF isole la relation directe entre une observation et son retard, en excluant l'influence de leurs retards intermédiaires.
- **Utilisation :** La PACF est principalement utilisée pour identifier l'ordre des processus autorégressifs (AR). Elle aide à déterminer le nombre de retards à utiliser dans un modèle AR en montrant l'étendue des corrélations directes à différents retards.

### Question 34 : Explain the principle of "parsimony" and its relationship with Ockham’s razor using the "theory of information criterion".

Le rasoir d'Ockham est un principe philosophique attribué à Guillaume d'Ockham, un frère franciscain et philosophe du XIVe siècle. Ce principe stipule que **"les entités ne doivent pas être multipliées au-delà de la nécessité"** ou, formulé plus simplement, **"la solution la plus simple est généralement la meilleure"**. En économétrie, cela signifie qu'une théorie avec moins de variables et d'hypothèses supplémentaires est préférée si elle explique les données aussi bien que des modèles plus complexes.

**Critères d'information en statistique**

En statistique, les critères d'information tels que le critère d'information d'Akaike (AIC) et le critère d'information bayésien (BIC) sont des outils utilisés pour appliquer le principe de parcimonie lors de la sélection de modèles. Ces critères évaluent non seulement la qualité de l'ajustement d'un modèle aux données, mais aussi la complexité du modèle, en pénalisant les modèles qui utilisent un plus grand nombre de paramètres.

**AIC (Akaike Information Criterion)**

L'AIC est calculé comme suit : $AIC= 2k−2\ln(L)$ où $k$ est le nombre de paramètres dans le modèle et $L$ la vraisemblance maximale du modèle. Un modèle avec un AIC plus faible est généralement préféré, car cela indique un meilleur équilibre entre la complexité du modèle et son ajustement aux données.

**BIC (Bayesian Information Criterion)**

Le BIC est similaire à l'AIC mais inclut un terme de pénalisation plus fort pour le nombre de paramètres : $BIC=ln(n)k−2ln(L)$
où $n$ est le nombre d'observations. Le BIC est donc plus strict que l'AIC en ce qui concerne la pénalisation des modèles plus complexes, surtout avec de grands échantillons.

En résumé, le principe de parcimonie et le rasoir d'Ockham nous encouragent à choisir des modèles plus simples lorsqu'ils sont suffisants pour expliquer les données, évitant ainsi la surcomplexité inutile. Les critères d'information comme l'AIC et le BIC sont des outils pratiques pour appliquer ce principe en évaluant et en comparant la simplicité relative et l'efficacité des différents modèles statistiques.

### Question 35 : Explain the problem of "auto-correlation of the errors".

L'autocorrélation des erreurs est un problème reccurent dans l'analyse de régression, particulièrement lorsqu'on a à faire à des séries temporelles ou des données spatiales. Ce phénomène se produit lorsque les résidus (erreurs) d'un modèle statistique ne sont pas indépendants les uns des autres, mais présentent au contraire **une corrélation à travers le temps ou l'espace**.

**Pourquoi l'autocorrélation des erreurs est-elle problématique ?**

**Invalidation des hypothèses statistiques :** La plupart des techniques de régression classiques supposent que les erreurs sont indépendantes et identiquement distribuées (i.i.d.). L'autocorrélation viole cette hypothèse, ce qui peut entraîner des estimations biaisées des paramètres du modèle et des erreurs types, rendant ainsi les tests statistiques (comme les tests de significativité des coefficients) incorrects.

**Efficacité des estimations :** L'autocorrélation peut réduire l'efficacité des estimations des paramètres du modèle, ce qui signifie que les estimations ne sont pas aussi précises qu'elles pourraient l'être si les erreurs étaient indépendantes. Cela peut mener à des conclusions erronées concernant les relations entre les variables.

Le **Test de Durbin-Watson** et l'analyse du **graphique de résidus** permettent de détecter l'autocorrélation des erreurs. Bien d'autres méthodes existent.

**Comment traiter l'autocorrélation des erreurs**

Si l'autocorrélation est détectée, plusieurs approches peuvent être prises pour y remédier :

**Utilisation de modèles appropriés :** Pour les données temporelles, des modèles comme les ARMA (AutoRegressive Moving Average) ou ARIMA (AutoRegressive Integrated Moving Average) peuvent être utilisés car ils incorporent explicitement l'autocorrélation dans le modèle.

**Transformation des données :** Des transformations comme la différenciation des données peuvent aider à éliminer l'autocorrélation en rendant les données stationnaires.

**Régression robuste :** L'application de techniques de régression robuste qui ajustent les erreurs types pour tenir compte de l'autocorrélation.

### Question 41 : Divide the sample in two groups: 1900-1960 and 1961-2000. Test the stability of coefficients between the two periods.

```python
import pandas as pd
import statsmodels.api as sm
from scipy import stats

# On effectue une fonction "perform_chow_test"
def perform_chow_test(df, before_year, after_year, y_var, x_vars):
    # Diviser les données en deux périodes
    df_before = df[df['year'] <= before_year]
    df_after = df[df['year'] > after_year]

    # Construire les modèles de régression pour chaque période
    X_before = sm.add_constant(df_before[x_vars])
    y_before = df_before[y_var]
    X_after = sm.add_constant(df_after[x_vars])
    y_after = df_after[y_var]

    # Ajuster les modèles
    model_before = sm.OLS(y_before, X_before).fit()
    model_after = sm.OLS(y_after, X_after).fit()

    # Combine both periods
    X_full = pd.concat([X_before, X_after])
    y_full = pd.concat([y_before, y_after])
    model_full = sm.OLS(y_full, X_full).fit()

    # Calculer le F-statistique de Chow
    numerator = (model_full.ssr - (model_before.ssr + model_after.ssr)) / model_full.df_model
    denominator = (model_before.ssr + model_after.ssr) / model_full.df_resid
    F = numerator / denominator

    # Obtenir la p-value
    p_value = stats.f.sf(F, model_full.df_model, model_full.df_resid)

    return F, p_value

# Nettoyage des données
df_filtre['unemployment_rate'] = pd.to_numeric(df_filtre['unemployment_rate'], errors='coerce')
df_filtre['consumer_price_inflation'] = pd.to_numeric(df_filtre['consumer_price_inflation'], errors='coerce')
df_filtre['real_gdp'] = pd.to_numeric(df_filtre['real_gdp'], errors='coerce')
df_filtre = df_filtre.dropna(subset=['unemployment_rate', 'consumer_price_inflation', 'real_gdp'])

# Effectuer le test de Chow
F_statistic, p_value = perform_chow_test(
    df=df_filtre, 
    before_year=1960, 
    after_year=1960, 
    y_var='real_gdp', 
    x_vars=['unemployment_rate', 'consumer_price_inflation']
)

# Afficher les résultats
print(f'Chow F-statistic: {F_statistic}')
print(f'Chow p-value: {p_value}')
alpha = 0.01
print(f'{"Les coefficients ne sont pas stables" if p_value < alpha else "Les coefficients sont stables"} entre les périodes (p < {alpha}).')
```

### Question 42 : Test the structural breakpoint using a trim ratio of 30% at the 1% level

Le ratio de rognage de 30% signifie que nous voulons exclure les 30% premières et dernières données de l'échantillon pour éviter les problèmes associés aux estimations aux extrémités de l'échantillon.

Le test CUSUM de stabilité des paramètres est utilisé pour détecter les ruptures structurelles dans une série temporelle.

```python
import pandas as pd
import statsmodels.api as sm
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.stats.diagnostic import recursive_olsresiduals

# Préparer les données
df_filtre['unemployment_rate'] = pd.to_numeric(df_filtre['unemployment_rate'], errors='coerce')
df_filtre['consumer_price_inflation'] = pd.to_numeric(df_filtre['consumer_price_inflation'], errors='coerce')
df_filtre['real_gdp'] = pd.to_numeric(df_filtre['real_gdp'], errors='coerce')
df_filtre = df_filtre.dropna().sort_values('year')

# Définir les limites pour le rognage
cut_off = int(len(df_filtre) * 0.3)
trimmed_df = df_filtre[cut_off:-cut_off]

# Construire le modèle
X_trimmed = sm.add_constant(trimmed_df[['unemployment_rate', 'consumer_price_inflation']])
y_trimmed = trimmed_df['real_gdp']

# Modèle de régression OLS
model = sm.OLS(y_trimmed, X_trimmed).fit()

# Obtenir les résidus récursifs
resids, cusum, cusumsq = recursive_olsresiduals(model)[-3:]

# Visualiser le test CUSUM
fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(cusum, label='CUSUM')
ax.axhline(y=0, color='r', linestyle='--')
ax.set_title('CUSUM for Parameter Stability with 30% Trimmed Data')
ax.legend()

# Calculer les seuils pour le test au niveau 1%
critical_value = 1.96 / np.sqrt(len(y_trimmed))  # Seuil approximatif pour 1%
ax.axhline(y=critical_value, color='g', linestyle='--', label='Critical Line (+)')
ax.axhline(y=-critical_value, color='g', linestyle='--', label='Critical Line (-)')

plt.show()

# Interprétation
if np.any(cusum > critical_value) or np.any(cusum < -critical_value):
    print("Structural break detected at the 1% level.")
else:
    print("No structural break detected at the 1% level.")
```

Dans le test CUSUM,  :

La ligne bleue représente les valeurs CUSUM calculées à partir des résidus du modèle. Un mouvement progressif et continu loin de zéro suggère des changements potentiels dans la relation entre les variables.

Les lignes horizontales pointillées vertes représentent les seuils critiques du test. Ces seuils sont déterminés par le niveau de signification qu'on a choisi (1% dans ce cas), et ils sont généralement basés sur des tables de distribution ou calculés par des formules spécifiques au test CUSUM.

L'interprétation générale est la suivante :

Si la ligne CUSUM franchit l'une de ces lignes de seuil (ce qui semble être le cas ici), cela indique qu'il y a eu un changement structurel significatif dans les données au niveau de confiance choisi (1% ici).

Si la ligne CUSUM reste à l'intérieur des lignes de seuil, cela suggérerait qu'il n'y a pas de preuve d'un changement structurel significatif.

Dans notre cas de figure, la ligne CUSUM monte fortement et franchit nettement les lignes de seuil, indiquant qu'il est très probable qu'il y ait eu un ou plusieurs changements structurels dans la relation entre les variables étudiées pendant la période analysée. Cela pourrait indiquer la nécessité d'un modèle différent ou d'une stratégie de modélisation différente pour les périodes avant et après le point de rupture détecté.
